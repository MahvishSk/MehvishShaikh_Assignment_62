### Topics:
**Functions**

```python
# 1️. Write a program using function to find greatest of three numbers.

# 2. Write a Python program using function to convert Celsius to Fahrenheit.

# 3. Write a Python function to remove a given word from a list and strip it at the same time.
```

```python
def greatest(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
z = int(input("Enter third number: "))

print("Greatest number is:", greatest(x, y, z))
```

```python
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

c = float(input("Enter temperature in Celsius: "))
print("Temperature in Fahrenheit:", celsius_to_fahrenheit(c))
```

```python
def remove_and_strip(word_list, word):
    new_list = []
    for item in word_list:
        if item.strip() != word:
            new_list.append(item.strip())
    return new_list

l = ["  apple ", " banana ", "  mango ", " banana "]
word = input("Enter word to remove: ")

print(remove_and_strip(l, word))
```
